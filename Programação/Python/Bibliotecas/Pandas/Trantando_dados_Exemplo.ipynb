```python
import pandas as pd

df = pd.read_csv('Horarios.csv', sep=';', encoding='cp1252')  # Padrão do Excel Windows
df.loc[1:5,:'horario']
```

```python
contagem = df['razao_social'].value_counts() == 2
lista_filtrada = list(contagem[contagem].index)
```

```python
resp = df[df['razao_social'].isin(lista_filtrada)].loc[::,:'horario']
resp
```

```python
resp['time'] = pd.to_datetime(resp['horario'])
resp
```

```python
Serie_max = pd.pivot_table(resp, index= 'razao_social', values=['time'], aggfunc= 'max')
Serie_min = pd.pivot_table(resp, index= 'razao_social', values=['time'], aggfunc= 'min')
Serie_max
```

```python
dif = Serie_max - Serie_min
dif
```
